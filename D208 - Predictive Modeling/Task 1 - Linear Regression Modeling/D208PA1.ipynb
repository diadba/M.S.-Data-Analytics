## Method Justification <a name="Method-Justification"></a>

### B1: Summary of Assumptions <a name="Summary-of-Assumptions"></a>

**Multiple Linear Regression**: Multiple linear regression is a useful statistical method that will assists in our analysis, by modeling the relationship between a continuous dependent variable, TotalCharge, and a range of independent variables by logical assumption of demographic characteristics, medical conditions and hospital services variables provided. This approach allows us to quantify (see A2) the impact, of these variables and influence on the daily charge of inpatient care, which may assist in the development of methods to mitigate readmission rates. However, the validity of this method relies on several significant assumptions.

**Assumptions**:

##### Linear Relationship &dagger;&dagger;
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In multiple linear regression, we assume that the dependent variable can be predicted from a linear combination of the independent variables. This assumption hinges on the principle that any change in an independent variable will produce a consistent/proportional change in the expected value of the dependent variable. For example, if we negate practicality only to explain the assumption then, with demographic characteristic Age as an independent variable, a linear relationship we would expect to see a consistent pattern in its effect on the daily charge of inpatient care, usually visualized in a scatterplot. If these relationships are linear, the variables can be included in a multiple linear regression model. Should the relationships be non-linear, alternative methods, such as variable transformations may be necessary.

##### No Multicollinearity &dagger;&dagger;
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Another assumption of multiple linear regression is no (or minimal) multicollinearity among the independent variables, meaning that each independent variable (predictor) should not be highly correlated with others. High correlation between predictors, such as HighBlood and Stroke, can lead to multicollinearity where stroke is often an unfortunate product of high blood pressure, which may obscure the distinct contribution of each independent variable, Stroke and HighBlood, and result in unstable estimates of the regression coefficients.

##### Homoscedasticity &dagger;&dagger;
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Homoscedasticity is an assumption in multiple linear regression that the variance of residuals, the differences between observed and predicted values by the model, remains constant across the prediction. This assumption means that as we move across different values predicted by the model, which are influenced by change in independent variables like Age or Income, the spread of the residuals should not increase or decrease. For our dependent variable, TotalCharge, we anticipate that the variance in the residuals is consistent regardless of whether the patient is young or old, less wealthy or more wealthy, etc. If the residuals exhibit patterns such as widening or narrowing with different predicted values, it suggests that the model may not appropriately capture the data's variation.

##### Correct Specification of the Model *
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Finally, the model is assumed to be correctly specified, which means that it includes all variables that are relevant to the outcome of interest/question (see A1) and accurately captures the true relationship between the dependent and independent variables. Additionally, it assumes that the model is appropriate for the data and that the relationship specified in the model correctly shows relationships among the variables. For example, if Age and TotalCharge are not linear then the model does not capture a relationship fit for this analysis.

### C4: Data Transformation <a name="Data-Transformation"></a>

**Goals**: The data transformation goals are related to the goals of the research question. In this step it is good practice to review the research question, assess the distributions, determine the transformations and implement them if necessary, lastly, rename variables for ease of access in model creation. The research question asks among different types of variables, which influence the TotalCharge value for the observations. The variables distributions can be found in C3. To best assess our data, if a variable shows a skewed univariate distribution, it may be best to apply a log or square root transformation. A new DataFrame, df_enc, will store the transformed data. We will also transform the data where necessary for categorical variables by use of dummy variables into a separate DataFrame, to preserve the original DataFrame's authenticity.

### D2: Justification of Model Reduction <a name="Justification-of-Model-Reduction"></a>

**Justification**: In this step, the goal is to justify the available methods to produce a reduced model that removes features that do not contribute significantly to the model or violates our assumptions in B1 (i.e., multicollinearity). In the following steps, the detection of VIF (Variance Inflation Factor) will evaluate multicollinearity in each variable. We will remove the values with a high VIF, now it is argued that a high VIF value is anywhere from above 2.5 to above 10 and sometimes lower. In this reduction we will remove variables with a VIF higher than 5 if any exist. VIF is great for diagnosing multicollinearity but it does not consider the statistical significance of the independent variables and TotalCharge. Backward elimination will remove the not statistically significant predictors (>0.05 p-values) as they are each described from the initial model summary.

### D3: Reduced Linear Regression Model <a name="Reduced-Linear-Regression-Model"></a>

**VIF**:

```python
# Create DataFrame to show column and respective VIF.
df_vif = pd.DataFrame()

# Column Named "Column Name"!
df_vif['Column Name'] = X.columns
# And their respective VIF Scores.
df_vif['VIF Score'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Output the VIF DataFrame.
df_vif
```

**NOTE**: No column is close to a high VIF score, we will try a different reduction method...

**Backward Elimination**: <br />
- Identify the independent variables to remove.
- Remove the variable with the highest p-value.
- Refit the model per removal.

**Identify the independent variables to remove**:

```python
# Get list of p-values.
p_values = initial_model.pvalues[1:] 
p_values_df = pd.DataFrame({'Column Name': p_values.index, 'p_value': p_values.values})

# Which columns are not statistically significant?
p_values_df[p_values_df['p_value'] > 0.05]
```

**Remove the variable with the highest p-value** & **Refit the model per removal**:

```python
# dependent variable.
y = df_enc['TotalCharge']

# Independent variables.
X = df_enc[model_cols].astype(int)
X = sm.add_constant(X)

#Remove until no more values > 0.05.
while True:
    # Recursive model fitting.
    wip_model = sm.OLS(y, X).fit()

    # Get new list of p-values.
    p_values = wip_model.pvalues.iloc[1:]

    #Highest p-value.
    max = p_values.max()

    if max > 0.05:
        # Variable to remove.
        ivar = p_values.idxmax()

        # Remove from DataFrame.
        X = X.drop(ivar, axis = 1)
    else:
        break

# Create reduced model from statsmodels Ordinary Least Squares linear reg.
reduced_model = sm.OLS(y, X).fit()
```
